```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
columnss=['t_dat','product_type_name','product_group_name','index_group_name']
enhanced_transactions_data_path="kaggle_data/enhanced_transactions.csv"
enhanced_transactions_are_go=pd.read_csv(enhanced_transactions_data_path,usecols = columnss)

```

```python
enhanced_transactions_are_go=enhanced_transactions_are_go.dropna(subset=['t_dat'])
enhanced_transactions_are_go['t_dat']=enhanced_transactions_are_go['t_dat'].astype('datetime64[ns]')
```

```python
enhanced_transactions_are_go.dtypes
```

```python
# thingy=enhanced_transactions_are_go.groupby([pd.Grouper(key = 't_dat', freq = 'M'),'product_type_name']) 

thingy=(
        enhanced_transactions_are_go
        .groupby([enhanced_transactions_are_go['t_dat'].dt.month,'product_type_name']) 
        .agg({'product_type_name':'count'})
        .rename(columns={'product_type_name':'orders'})
        .reset_index()
)
```

```python
thingy.head()
```

```python
other_thingy=(
        thingy
        .groupby(['product_type_name']) 
        .agg({'orders':'sum'})
        .reset_index()
)
other_thingy.sort_values(by=['orders'],ascending=False).head(20)
```
